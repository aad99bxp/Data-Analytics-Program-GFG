# **String Operations**

## Check whether is a given substring or not

```python
#1. Check whether is a given substring or not
s1= "Abhayadityadubey"
s2="aditya"
print(s2 in s1)
```

## String Concatenation

```python
print(s1+s2)
```

# Finding position of the first character of a substring

```python
# USING INDEX - Gives the position of the first string
print(s1.index(s2))
```

```python
# USING RINDEX - Gives the position where the substring ends
a='tyequov'
b='equo'
print(a.rindex(b))
```

```python
# Providing paramters in the Index Function
print(s1.index(s2, 6, 10)) #It will find the substring from the position 6 to 10
# If the string is not present, it will throw error
```

```python
# Thus, find fucntion can be used, it works the same way like index. The only difference is it will not throw error if the substring is not found but return -1.
print(s1.find(s2))
print(s1.find(s2, 6, 10))
```

# OTHER OPEARTIONS

```python
str= 'Abhay aditya'
str2= str.lower()
str3= str.upper()
print(str2)
print(str3)
print(str2.islower())
print(str3.isupper())
print(len(str))
print(str.title()) # converts string to Title Case
```

```python
print(str.startswith('Abh'))
print(str.endswith('y'))
```

```python
# Startswith and Endswith take two additional params- the starting index and the final index
print(str.startswith('y', 2, 4))
print(str.endswith('y', 2, 4))
```

```python
x= "Abhay, Aditya, Dubey" # Split functions splits the string on the basis of delimiter and returns a list
x.split()
```

```python
x.split(",")
```

```python
# The opposite of Split fucntion is Join Function, which takes join parameter and a list as input and returns a string
y= ['Abhay', 'Aditya', 'dubey']
print(" ".join(y))
print(",".join(y))
```

```python
# STRIP FUCNTIONS- These functions are used to remove unwanted characters from the corners of string
name="--Ab--hay--"
print(name.strip("-"))
print(name.lstrip("-"))
print(name.rstrip("-"))
```

# **Pattern Searching**

```python
txt=input("Input Text")
pattern=input("Pattern to be searched")
if pattern in txt:
  print("Pattern Found")
else:
  print("Pattern Not Found")
```

```python
txt=input("Input Text")
pattern=input("Pattern to be searched")
pos= txt.find(pattern)
while pos>=0:
  print("Pattern Found", pos)
  pos = txt.find(pattern, pos+1)
```

# **String Slicing**

```python
# Python Program to
# demonstrate String slicing

# Creating a String
String1 = "GeeksForGeeks"
print("Initial String: ")
print(String1)

# Printing 3rd to 12th character
print("\nSlicing characters from 3-12: ")
print(String1[3:12])

# Printing characters between
# 3rd and 2nd last character
print("\nSlicing characters between " +
      "3rd and 2nd last character: ")
print(String1[3:-2])
```
